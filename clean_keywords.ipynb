```python
# import dependencies
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
# Read the csv file to dataframe
movies_metadata_df = pd.read_csv("movies_metadata.csv")
movies_metadata_df.head()
```

### Drop unecessary rows and columns

```python
# Drop rows with zero revenue
movies_df =movies_metadata_df[movies_metadata_df['revenue'] !=0]
```

```python
len(movies_df.index)
```

```python
# drop duplicate by a column name
movies_df = movies_df.drop_duplicates(['id'])
```

```python
len(movies_df.index)
```

```python
# Drop imdb_id column
movies_df = movies_df.drop(['imdb_id'], axis=1)
```

```python
# Drop homepage
movies_df = movies_df.drop(['homepage'], axis=1)
```

```python
# Drop poster_path
movies_df = movies_df.drop(['poster_path'], axis=1)
```

```python
# Drop overview
movies_df = movies_df.drop(['overview'], axis=1)
```

```python
# Convert releast_date to datetime using ‘coerce’ to set invalid parsing as NaT.
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
```

```python
movies_df = movies_df.dropna(subset=['release_date'])
```

```python
len(movies_df.index)
```

```python
movies_df.head()
```

```python
###  Clean Keywords File

# Read the csv file to dataframe
keywords_df = pd.read_csv("keywords.csv")
keywords_df.head()
```

```python

len(keywords_df.index)
```

```python
keywords_df.info()
```

```python
# Convert id to string type
data_types_dict = {'id': str} 
```

```python
keywords_df = keywords_df.astype(data_types_dict) 

keywords_df.head()
```

```python
# Drop NaNs
keywords_df = keywords_df.dropna()
len(keywords_df.index)
```

```python
# Get just the 'name' key values - result is a list of just the values
keywords_df['keywords'] = keywords_df['keywords'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
len(keywords_df.index)
```

```python
keywords_df.head()
```

```python
# Join keywords dataframe with movies_df DataFrame to get only keywords with matching movie 
keywords_df = pd.merge(keywords_df, movies_df['id'], how='inner', left_on=['id'], right_on=['id'])
keywords_df.columns
```

```python
len(keywords_df.index)
```

```python
keywords_df.sample(50)
```

```python
# Get non-null count (there are no nulls!)
keywords_df.info(null_counts=True)
```

```python
# drop duplicates by a column name
keywords_df = keywords_df.drop_duplicates(['id'])
```

```python
len(keywords_df.index)
```

```python
# write clean keywords data to csv file for loading to db
keywords_df.to_csv('keywords_clean.csv', index=False)
```
