## Clean the movies_metatdata dataset and ratings dataset 
## prior to loading to database tables

```python
# import dependencies
import pandas as pd
import numpy as np
from ast import literal_eval
```

```python
# Read the csv file to dataframe
movies_metadata_df = pd.read_csv("movies_metadata.csv")
movies_metadata_df.head()
```

```python
len(movies_metadata_df.index)
```

### Drop unecessary rows and columns

```python
# Drop rows with zero revenue
movies_df =movies_metadata_df[movies_metadata_df['revenue'] !=0]
```

```python
len(movies_df.index)
```

```python
# drop duplicates by a column name
movies_df = movies_df.drop_duplicates(['id'])
```

```python
len(movies_df.index)
```

```python
# Drop imdb_id column
movies_df = movies_df.drop(['imdb_id'], axis=1)
```

```python
# Drop homepage
movies_df = movies_df.drop(['homepage'], axis=1)
```

```python
# Drop poster_path
movies_df = movies_df.drop(['poster_path'], axis=1)
```

```python
# Drop overview
movies_df = movies_df.drop(['overview'], axis=1)
```

```python
# Convert releast_date to datetime using ‘coerce’ to set invalid parsing as NaT.
movies_df['release_date'] = pd.to_datetime(movies_df['release_date'], errors='coerce')
```

```python
movies_df = movies_df.dropna(subset=['release_date'])
```

```python
len(movies_df.index)
```

```python
movies_df.head()
```

### Convert "belongs_to_column" from list type to boolean

```python
# Put "belongs_to_column" into a series
collection_col = pd.Series(movies_df.belongs_to_collection)
```

```python
# Drop the "belongs_to_column" from dataframe
movies_df = movies_df.drop(['belongs_to_collection'], axis=1)
```

```python
movies_df.head()
```

```python
# Drop the Nan rows from Series
collection_col = collection_col.dropna()
```

```python
# Change data type to Boolean
collection_col = collection_col.astype('bool')
```

```python
collection_col
```

```python
# Convert Series to DataFrame
collection_col = collection_col.to_frame()
```

```python
# Join collections column back into movies_df DataFrame
movies_df = movies_df.join(collection_col)
```

```python
# Fill remaining NaN with False
movies_df["belongs_to_collection"].fillna("False", inplace=True)
```

```python
movies_df.head(10)
```

```python
len(movies_df.index)
```

## Separate "production_companies" column (a list of dictionaries) to

### Convert "spoken_languages" from list of dictionaries to just the languages count

```python
# Put "spoken_languages" into a series
spoken_col = pd.Series(movies_df.spoken_languages)
```

```python
# Drop the "spoken_languages" from dataframe
movies_df = movies_df.drop(['spoken_languages'], axis=1)
```

```python
# Convert Series to DataFrame
spoken_df = pd.DataFrame(spoken_col)
spoken_df.head(10)
```

```python
# Get just the 'iso_639_1' key values - result is a list of just the language abbreviation values
spoken_df['spoken_languages'] = spoken_df['spoken_languages'].fillna('[]').apply(literal_eval).apply(lambda x: [i['iso_639_1'] for i in x] if isinstance(x, list) else [])
```

```python
spoken_df.head(10)
```

```python
spoken_df['spoken_languages'].value_counts()
```

```python
for index, row in spoken_df.iterrows():
    language_count = 0
    if len(row['spoken_languages'])==0:
        spoken_df.loc[index,'spoken_languages'] = language_count
    else:
        for language in row['spoken_languages']:
            language_count += 1
            #print(language_count)
            spoken_df.loc[index,'spoken_languages'] = language_count
```

```python
spoken_df.tail(50)
```

```python
# Join new spoken_languages column (which is now a count) back into movies_df DataFrame
movies_df = movies_df.join(spoken_df)
```

```python
movies_df.columns
```

## separate columns for each genre
## Write file with genre columns to load to genres table

```python
# Put "genres" into a series
genres_col = pd.Series(movies_df.genres)
```

```python
# Drop the "genres" from dataframe
movies_df = movies_df.drop(['genres'], axis=1)
```

```python
# Drop any Nan rows from Series - there are none
genres_col = genres_col.dropna()
```

```python
genres_col
```

```python
# Convert Series to DataFrame
genres_df = pd.DataFrame(genres_col)
genres_df.head(10)
```

```python
len(genres_df)
```

```python
# Get just the 'name' key values - result is a list of just the values
genres_df['genres'] = genres_df['genres'].fillna('[]').apply(literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
genres_df.head(10)
```

```python
# Add genre columns

genres_df["drama"] = False
genres_df["comedy"] = False
genres_df["thriller"] = False
genres_df["romance"] = False
genres_df["action"] = False
genres_df["horror"] = False
genres_df["crime"] = False
genres_df["documentary"] = False
genres_df["adventure"] = False
genres_df["science_fiction"] = False
genres_df["family"] = False
genres_df["mystery"] = False
genres_df["fantasy"] = False
genres_df["animation"] = False
genres_df["foreign"] = False
genres_df["music"] = False
genres_df["history"] = False
genres_df["war"] = False
genres_df["western"] = False
genres_df["tv_movie"] = False
genres_df["others"] = False
```

```python
genres_df.head(10)
```

```python
# Nested Loops. Loop through each row of the dataframe. The data type of the column 'genres' is a list of genres 
# so loop through each element of that list.  This gets the value of the element. Match the value to the appropriate column.
for index, row in genres_df.iterrows():
    print(row['genres'])
    for genre in row['genres']:
        print(genre)
        if genre == "Drama": 
            genres_df.loc[index, "drama"] = True
        elif  genre == "Comedy":
            genres_df.loc[index, "comedy"] = True
        elif genre == "Thriller":
            genres_df.loc[index, "thriller"] = True
        elif genre == "Romance":
            genres_df.loc[index, "romance"] = True
        elif genre == "Action":
            genres_df.loc[index, "action"] = True
        elif genre == "Horror":
            genres_df.loc[index, "horror"] = True
        elif genre == "Crime":
            genres_df.loc[index, "crime"] = True
        elif genre == "Documentary":
            genres_df.loc[index, "documentary"] = True
        elif genre == "Adventure":
            genres_df.loc[index, "adventure"] = True
        elif genre == "Science Fiction":
            genres_df.loc[index ,"science_fiction"] = True
        elif genre == "Family":
            genres_df.loc[index, "family"] = True
        elif genre == "Mystery":
            genres_df.loc[index, "mystery"] = True
        elif genre == "Fantasy":
            genres_df.loc[index, "fantasy"] = True
        elif genre == "Animation":
            genres_df.loc[index, "animation"] = True
        elif genre == "Foreign":
            genres_df.loc[index, "foreign"] = True
        elif genre == "Music":
            genres_df.loc[index, "music"] = True
        elif genre == "History":
            genres_df.loc[index, "history"] = True
        elif genre == "War":
            genres_df.loc[index, "war"] = True
        elif genre == "Western":
            genres_df.loc[index, "western"] = True
        elif genre == "TV Movie":
            genres_df.loc[index, "tv_movie"] = True
        else:
            genres_df.loc[index,"others"] = True
```

###  Clean Ratings File

```python
# Read the csv file to dataframe
user_ratings_df = pd.read_csv("ratings.csv")
user_ratings_df.head()
```

```python
len(user_ratings_df.index)
```

```python
# Convert movieId to string type
data_types_dict = {'movieId': str} 

user_ratings_df = user_ratings_df.astype(data_types_dict) 
```

```python
user_ratings_df.head()
```

```python
# Convert timestamp column to datetime data type
user_ratings_df['timestamp'] = pd.to_datetime(user_ratings_df['timestamp'], unit='s')
```

```python
user_ratings_df.sample(10)
```

```python
# Drop NaNs
user_ratings_df = user_ratings_df.dropna()
```

```python
len(user_ratings_df.index)
```

```python
# Merge two dataframe to get only the ratings that have a corresponding movie in movies_df
movies_ratings_df = pd.merge(user_ratings_df, movies_df['id'], how='inner', left_on=['movieId'], right_on=['id'])
```

```python
len(movies_ratings_df.index)
```

```python
# Drop movieId column
movies_ratings_df = movies_ratings_df.drop(['movieId'], axis=1)
```

```python
# Reorder columns
movies_ratings_df = movies_ratings_df[["userId","id", "rating", "timestamp"]]
```

```python
len(movies_ratings_df.index)
```

```python
movies_ratings_df.head()
```

```python
# Get non-null count (there are no nulls!)
movies_ratings_df.info(null_counts=True)
```

```python
# write clean ratings data to csv file for loading to db
movies_ratings_df.to_csv('ratings_clean.csv', index=False)
```
